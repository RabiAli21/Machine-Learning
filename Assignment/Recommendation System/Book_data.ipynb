```python
data1 = data.drop_duplicates(subset=["User.ID"])
user_book_df = data1.pivot(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index()
```

```python
user_book_df
```

```python
user_book_df.index = data1["User.ID"].unique()
```

```python
user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
df_user_sim = pd.DataFrame(user_sim)
```

```python
df_user_sim.index = data1["User.ID"].unique()
df_user_sim.columns = data1["User.ID"].unique()
```
